```python
from ultralytics import YOLO
import os
```

```python
size = 0

path = '/Users/miguel/GitHub/bone_tumour_classification/test_images'
for file in os.listdir(path):
    if file.endswith('.png'):
        size += os.path.getsize(os.path.join(path, file))
```

```python
model = YOLO('models/best.pt')

for i in range(size):
    results = model(f'test_image{i}.png', save=True)
guess = results[0].verbose().strip()
print(guess)
```
